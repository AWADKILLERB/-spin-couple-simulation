```python
import os

# List all files in the current directory
files_in_dir = os.listdir('.')

# Filter for files with the '.png.png' extension
png_png_files = [f for f in files_in_dir if f.endswith('.png.png')]

print(f"Found {len(png_png_files)} files with double .png extension:")
for old_name in png_png_files:
    new_name = old_name.replace('.png.png', '.png')
    os.rename(old_name, new_name)
    print(f"Renamed '{old_name}' to '{new_name}'")

print("File renaming complete.")
```

Now that the image files are renamed, I will create the `pictures` directory and move the image files into it.

```python
import os

pictures_dir = 'pictures'
if not os.path.exists(pictures_dir):
    os.makedirs(pictures_dir)
    print(f"Created directory: {pictures_dir}")

image_files = ['figure1.png', 'figure2.png', 'figure3.png']

for file_name in image_files:
    if os.path.exists(file_name):
        os.rename(file_name, os.path.join(pictures_dir, file_name))
        print(f"Moved '{file_name}' to '{pictures_dir}/{file_name}'")
    else:
        print(f"Warning: File '{file_name}' not found in current directory.")

print("Image files moved to 'pictures' directory.")
```
